```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('https://raw.githubusercontent.com/huynhhoc/phantichdulieu/main/Data/babies.txt',sep='\s+')
print(data)
```

```python
df_cohutthuoc=pd.DataFrame(data,columns=['bwt','smoke']).query('smoke==1')
df_khonghutthuoc=pd.DataFrame(data,columns=['bwt','smoke']).query('smoke==0')
print('Me co hut thuoc')
print(df_cohutthuoc.describe())
print('Var:',df_cohutthuoc['bwt'].var(skipna=True))
print('skew:',df_cohutthuoc['bwt'].skew())
print('kurtosis:',df_cohutthuoc['bwt'].kurtosis()+3)
```

```python
print('Me khong hut thuoc')
print(df_khonghutthuoc.describe())
print('Var:',df_khonghutthuoc['bwt'].var(skipna=True))
```

```python
plt.hist(df_cohutthuoc['bwt'],bins=12)
plt.show()
```

```python
plt.boxplot(df_cohutthuoc['bwt'])
plt.show()
```

```python
data.boxplot(by='smoke', column='bwt')
plt.show()
```

```python
from statsmodels.graphics.gofplots import qqplot 
qqplot(df_cohutthuoc['bwt'],line='s')
qqplot(df_khonghutthuoc['bwt'],line='s')
plt.show()
```
